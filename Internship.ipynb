### 1. Outliers have been detected and they need to be removed
### 2. Most columns are skewed so log transform is used

```python
skewed_features = ['Adult_mortality', 'GDP_per_capita', 'Infant_deaths', 'Under_five_deaths', 'Population_mln']
# Applying transformation to Skewed columns using Log transformation
for col in skewed_features:
    X_train[col] = np.log1p(X_train[col])
    X_test[col] = np.log1p(X_test[col])
```

```python
# Removing outliers using IQR method
def remove_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    
    # Filter the DataFrame to keep values within the bounds
    filtered_df = df[(df[column] >= lower_bound) & (df[column] <= upper_bound)]
    return filtered_df
```

```python
# Combining X_train and y_train to remove outliers consistently
train_data = pd.concat([X_train, y_train], axis=1)

# Printing initial shape before outlier removal
print("Shape of training data before outlier removal:", train_data.shape)

# Removing outliers from key features
for col in skewed_features + ['Life_expectancy']:
    train_data = remove_outliers_iqr(train_data, col)

# Separating back into X_train and y_train
X_train_clean = train_data.drop('Life_expectancy', axis=1)
y_train_clean = train_data['Life_expectancy']
train_data_cleaned = pd.concat([X_train_clean, y_train_clean], axis=1)
# Printing final shape after outlier removal
print("Shape of training data after outlier removal:", X_train_clean.shape)
```

```python
# Storing the original data for plot comparision
X_train_original = X_train.copy()
train_data_original = pd.concat([X_train.copy(), y_train.copy()], axis=1)
```

## Summary of Findings

### The dataset contains rich set of features but a significant portion of data needed cleaning and preparation before using them for any machine learning model
##### 1. Distributions and Outliers: Most of the key numerical features, such as Adult_mortality, GDP_per_capita, and Infant_deaths, are highly right-skewed. The Life_expectancy target variable is left-skewed. Box plots would reveal a large number of outliers in almost all of the skewed features, highlighting the need for careful handling.
##### 2. Correlations: Correlation heatmap showed several strong relationships, both positive and negative. For example, Life_expectancy is strongly negatively correlated with Adult_mortality and Infant_deaths and positively correlated with Schooling and GDP_per_capita. Many features, however, have very weak correlations with the target variable, which is important to consider for feature selection.
##### 3. Data Preparation: The data's characteristics indicate that a one-size-fits-all approach to model training is not suitable. Linear Regression would require significant preprocessing, including log transformation for skewed features and the removal of outliers. In contrast, Decision Trees and Random Forests are naturally robust to both skewed distributions and outliers, making them a simpler choice in terms of data preparation.

## Key Takeaways

##### 1. To fit in linear model like Linear Regression and perform well, transformation of skewed features is needed and reduction of influence of outliers is needed.
##### 2. Model choice influences the Data Preparation methods.
##### 3. The significant number of outliers in this data would heavily impact a Linear Regression model's performance. It is important to decide whether to remove them or use a model that is inherently robust to their presence.

## Comparison of feature weights across models

### 1. Linear Regression
#### In Linear Regression, the "importance" of a feature is represented by its coefficient. A larger absolute value of a coefficient indicates a stronger relationship with the target variable, Life_expectancy. A positive coefficient means the feature is positively correlated, while a negative coefficient indicates a negative correlation.

#### Key Positive Predictors: Features like Schooling, Income_composition_of_resources, and BMI likely have the largest positive coefficients. This means that a higher value in these features is strongly associated with a higher life expectancy. In the real world, this makes perfect sense: better education and economic stability generally lead to improved health outcomes and a longer lifespan.

#### Key Negative Predictors: Adult_mortality and Incidents_HIV are expected to have the largest negative coefficients. A higher adult mortality rate or a higher incidence of HIV would be strongly associated with a lower life expectancy.

### 2. Decision Tree & Random Forest
#### These models determine feature importance differently, often using metrics like Gini impurity or information gain. Essentially, they measure how much a feature contributes to reducing the error of the model.

#### Top Features: For both Decision Tree and Random Forest, you'll likely see a few features dominating the importance rankings. The top features are consistently those that provide the most "bang for the buck" in a single split. In this dataset, the most important features are consistently Adult_mortality and Incidents_HIV.

#### Interpretation: The high importance of these features means they are the most effective at partitioning the data to predict Life_expectancy. A single split based on a high vs. low Adult_mortality value, for instance, is extremely effective at separating countries with low vs. high life expectancy. This aligns with real-world understanding: mortality rates are a direct measure of population health and are a primary determinant of lifespan.

# Week 3: Deep Learning

```python
from tensorflow import keras
from tensorflow.keras import layers

# Defining the Keras Sequential model
model = keras.Sequential([
    layers.Dense(64, activation='relu', input_shape=[X_train_clean.shape[1]]),
    layers.Dense(32, activation='relu'),
    layers.Dense(1) # Output layer for regression
])
```

```python
# Compiling the model
model.compile(
    optimizer='adam',
    loss='mean_squared_error',
    metrics=['mae', 'mse']
)
```

```python
# model summary
model.summary()
```

```python
# Fitting the model to the training data
history = model.fit(
    X_train_clean,
    y_train_clean,
    epochs=100,
    validation_split=0.2,
    verbose=1
)
```

```python
# Evaluate the model on the test data
loss, mae, mse = model.evaluate(X_test, y_test, verbose=0)
print(f"\nTest set MAE: {mae:.2f}")
print(f"Test set MSE: {mse:.2f}")
```

```python
# Plotting the training and validation loss
plt.figure(figsize=(10, 6))
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss (Mean Squared Error)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Defining the tuned Keras Sequential model with Dropout and Batch Normalization
model_tuned = keras.Sequential([
    layers.Dense(128, input_shape=[X_train_clean.shape[1]]),
    layers.BatchNormalization(), # Adding Batch Normalization
    layers.Activation('relu'),
    layers.Dropout(0.3), # Adding Dropout layer
    layers.Dense(64),
    layers.BatchNormalization(), # Adding Batch Normalization
    layers.Activation('relu'),
    layers.Dropout(0.3), # Adding Dropout layer
    layers.Dense(1) # Output layer for regression
])
```

```python
# Compiling the tuned model
model_tuned.compile(
    optimizer='adam',
    loss='mean_squared_error',
    metrics=['mae', 'mse']
)
```

```python
# Training the tuned model
history_tuned = model_tuned.fit(
    X_train,
    y_train,
    epochs=100,
    validation_split=0.2,
    verbose=0
)
```

```python
# Evaluating the new model on the test data
loss, mae, mse = model_tuned.evaluate(X_test, y_test, verbose=0)
print(f"\nNew Model Test set MAE: {mae:.2f}")
print(f"New Model Test set MSE: {mse:.2f}")
```

```python
# Plotting the training and validation loss
plt.figure(figsize=(10, 6))
plt.plot(history_tuned.history['loss'], label='Training Loss')
plt.plot(history_tuned.history['val_loss'], label='Validation Loss')
plt.title('Training and Validation Loss Over Epochs')
plt.xlabel('Epoch')
plt.ylabel('Loss (Mean Squared Error)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import streamlit as st

# Loading the trained model
try:
    model = joblib.load('random_forest_model.pkl')
except FileNotFoundError:
    st.error("Error: 'random_forest_model.pkl' not found. Please ensure the trained model file is in the same directory.")
    st.stop()

# Loading the original data to get the list of regions for one-hot encoding
try:
    df_original = pd.read_csv('Life_Expectancy_Data.csv')
    regions = sorted(df_original['Region'].unique())
except FileNotFoundError:
    st.error("Error: 'Life_Expectancy_Data.csv' not found. This is needed for the dashboard to function.")
    st.stop()

st.set_page_config(page_title="Life Expectancy Predictor", layout="wide")

# Dashboard Title and Description
st.title("Life Expectancy Predictor")
st.markdown("### Predict Life Expectancy based on various health and economic factors.")
st.write("Use the sidebar on the left to input the values for the different features.")

# Sidebar for user input
with st.sidebar:
    st.header("Input Features")
    st.markdown("Enter the values for the following predictors:")

    year = st.number_input("Year", min_value=2000, max_value=2015, value=2010)
    infant_deaths = st.number_input("Infant Deaths (per 1000 births)", min_value=0.0, value=20.0, step=0.1)
    under_five_deaths = st.number_input("Under Five Deaths (per 1000 births)", min_value=0.0, value=25.0, step=0.1)
    adult_mortality = st.number_input("Adult Mortality (per 1000 population)", min_value=0.0, value=150.0, step=0.1)
    alcohol_consumption = st.number_input("Alcohol Consumption (litres)", min_value=0.0, value=5.0, step=0.1)
    hepatitis_b = st.number_input("Hepatitis B (%)", min_value=0.0, max_value=100.0, value=80.0, step=0.1)
    measles = st.number_input("Measles (per 1000 population)", min_value=0.0, value=50.0, step=0.1)
    bmi = st.number_input("BMI", min_value=0.0, value=22.0, step=0.1)
    polio = st.number_input("Polio (%)", min_value=0.0, max_value=100.0, value=80.0, step=0.1)
    diphtheria = st.number_input("Diphtheria (%)", min_value=0.0, max_value=100.0, value=80.0, step=0.1)
    incidents_hiv = st.number_input("Incidents HIV (per 1000 population)", min_value=0.0, value=0.5, step=0.01)
    gdp_per_capita = st.number_input("GDP per Capita ($)", min_value=0.0, value=5000.0, step=100.0)
    population_mln = st.number_input("Population (millions)", min_value=0.0, value=50.0, step=0.1)
    thinness_ten_nineteen_years = st.number_input("Thinness 10-19 Years (%)", min_value=0.0, value=5.0, step=0.1)
    thinness_five_nine_years = st.number_input("Thinness 5-9 Years (%)", min_value=0.0, value=5.0, step=0.1)
    schooling = st.number_input("Schooling (years)", min_value=0.0, value=10.0, step=0.1)
    economy_status = st.radio("Economy Status", ["Developed", "Developing"])
    region = st.selectbox("Region", regions)

    predict_button = st.button("Predict Life Expectancy")

# Prediction logic
if predict_button:
    # Prepare the input data
    input_data = {
        'Year': year,
        'Infant_deaths': infant_deaths,
        'Under_five_deaths': under_five_deaths,
        'Adult_mortality': adult_mortality,
        'Alcohol_consumption': alcohol_consumption,
        'Hepatitis_B': hepatitis_b,
        'Measles': measles,
        'BMI': bmi,
        'Polio': polio,
        'Diphtheria': diphtheria,
        'Incidents_HIV': incidents_hiv,
        'GDP_per_capita': gdp_per_capita,
        'Population_mln': population_mln,
        'Thinness_ten_nineteen_years': thinness_ten_nineteen_years,
        'Thinness_five_nine_years': thinness_five_nine_years,
        'Schooling': schooling,
        'Economy_status_Developed': 1 if economy_status == "Developed" else 0,
        'Economy_status_Developing': 1 if economy_status == "Developing" else 0
    }
    
    # Creating DataFrame from input data
    input_df = pd.DataFrame([input_data])
    
    # Handling one-hot encoding for 'Region'
    for r in regions:
        input_df[f'Region_{r}'] = 1 if r == region else 0
    
    # Ensuring column order matches the training data
    ## This is critical for the model to make correct predictions
    feature_order = [
        'Year', 'Infant_deaths', 'Under_five_deaths', 'Adult_mortality',
        'Alcohol_consumption', 'Hepatitis_B', 'Measles', 'BMI', 'Polio',
        'Diphtheria', 'Incidents_HIV', 'GDP_per_capita', 'Population_mln',
        'Thinness_ten_nineteen_years', 'Thinness_five_nine_years', 'Schooling',
        'Economy_status_Developed', 'Economy_status_Developing'
    ] + [f'Region_{r}' for r in regions]

    input_df = input_df[feature_order]
    
    # Making the prediction
    prediction = model.predict(input_df)[0]
    
    # Displaying the prediction
    st.markdown("---")
    st.subheader("Prediction Result")
    st.metric(label="Predicted Life Expectancy", value=f"{prediction:.2f} years")
```
